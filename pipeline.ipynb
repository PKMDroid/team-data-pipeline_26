```python
#Data creation 
import pandas as pd
import numpy as np

sales_data = pd.DataFrame({
    "Product": ["Books", "Pens", "Bags", "Shoes", "Shirts"],
    "Quantity": [10, 25, 7, 5, 12],
    "Unit_Price": [5000, 1000, 30000, 80000, 25000]
})

sales_data["Revenue"] = sales_data["Quantity"] * sales_data["Unit_Price"]

sales_data
```

```python
# Descriptive statistics/average 

average_revenue = sales_data["Revenue"].mean()

print("Average Revenue:", average_revenue)

sales_data.describe()
```

# Data Pipeline Explanation

This notebook demonstrates a simple data pipeline.

First, the data is created using product names, quantities, and unit prices. The revenue for each product is calculated by multiplying the quantity by the unit price.

Next, the data is analyzed by calculating the average revenue and generating descriptive statistics. These statistics help summarize the dataset by showing values such as the mean, minimum, maximum, and spread of the numerical data.

Overall, this pipeline shows how raw data can be created, processed, analyzed, and explained in a clear and organized way.
